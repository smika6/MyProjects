##  TSP GA Functions
### Random Route Function

```python
def random_route(data):
    return random.sample(data, len(data))
```

### Function to Align Route to Start with Specified Point

```python
def align_route_starts_with_pt(route, pt):
    
    start = None
    for i,r in enumerate(route):
        if r == pt:
            start = i
    else:
        print('Point was not found in the route.')

    return route[start:] + route[0:start]
```

### Function for Euclidean Distance

```python
def euclidean_distance(a,b):
    return np.sqrt(  ( abs( b[0] - a[0] ) )**2 + ( abs( b[1] - a[1] ) )**2  )
```

### Equal Cost Route Distance Function

```python
def calculate_route_distance(route):
    
    route_distance = 0
    
    for i in range(0, len(route)):
        
        city_a = route[i]
        city_b = None
        
        if i + 1 < len(route):
            city_b = route[i + 1]
        else:
            city_b = route[0]
        
        route_distance += euclidean_distance(city_a,city_b)
    
    return route_distance
```

### Route Fitness Function

```python
def evaluate_route_fitness(route):
    return 1.0 / calculate_route_distance(route);
```

###  Function for Initialization of a Population

```python
def initialize_population(population_size, dataset):
    population = [] 

    for _ in range(population_size):
        population.append(random_route(dataset))
        
    population_with_distances_and_fitness = []

    for candidate in population:
        dst = calculate_route_distance(candidate)
        population_with_distances_and_fitness.append( (candidate, dst, 1/dst) )
    
    return population_with_distances_and_fitness
```

### Function to Sort Routes by Distance

```python
def rank_routes(population):
    return sorted( population, key = lambda candidate:candidate[2], reverse = True)
```

### Selection Function
Fitness proportionate selection (the version implemented below): The fitness of each individual relative to the population is used to assign a probability of selection. Think of this as the fitness-weighted probability of being selected.

```python
def selection(population_ranked, elite_count):
    mating_pool = []
    
    fitness = {'Fitness':[c[2] for c in population_ranked]}
    
    df = pd.DataFrame(fitness)
    df['cum_sum'] = df.Fitness.cumsum()
    df['cum_perc'] = 100 * df.cum_sum/df.Fitness.sum()
    
    for i in range(0,elite_count):
        mating_pool.append(population_ranked[i])
    
    for _ in range(0,len(population_ranked) - elite_count):
        pick = 100*random.random()
        for i in range(len(population_ranked)):
            if pick <= df.iat[i,2]:
                mating_pool.append(population_ranked[i])
                break
    
    return mating_pool
```

### Crossover Function

```python
def ordered_crossover(parent1, parent2):
    parent1_route = parent1[0]
    parent2_route = parent2[0]
    
    child = []
    gene = []
    
    geneA = int(random.random() * len(parent1_route))
    geneB = int(random.random() * len(parent1_route))
    
    startGene = min(geneA, geneB)
    endGene = max(geneA, geneB)
    
    for i in range(startGene, endGene):
        gene.append(parent1_route[i])
        
    remainder = [allele for allele in parent2_route if allele not in gene]
    
    child = gene + remainder
    
    d = calculate_route_distance(child)
    
    return (child,d,1.0/d)
```

```python
def breed_population(matingpool, elite_count):
    children = []
    pool = random.sample(matingpool, len(matingpool))

    for i in range(0,elite_count):
        children.append(matingpool[i])
    
    for i in range(0, len(matingpool) - elite_count ):
        child = ordered_crossover(pool[i], pool[len(matingpool)-i-1])
        children.append(child)
    
    return children
```

### Mutation Function

```python
def mutation(candidate, mutation_rate):
    mutant = candidate[0]
    for swapped in range(len(mutant)):
        if(random.random() < mutation_rate):
            swapWith = int(random.random() * len(mutant))
            
            city1 = mutant[swapped]
            city2 = mutant[swapWith]
            
            mutant[swapped] = city2
            mutant[swapWith] = city1
            
            break
    
    d = calculate_route_distance(mutant)
    return (mutant, d, 1/d)
```

```python
def mutate_population(population, mutation_rate):
    mutated_pop = []
    
    for p in population:
        mutant = mutation(p, mutation_rate)
        mutated_pop.append(mutant)
    
    return mutated_pop
```

### Transition Function

```python
def next_generation(current_generation, elite_count, mutation_rate):
    population_ranked = rank_routes(current_generation)
    matingpool = selection(population_ranked, elite_count)
    children = breed_population(matingpool, elite_count)
    next_generation = mutate_population(children, mutation_rate)
    return next_generation
```

```python
def next_generation_for_treads(current_generation, elite_count, mutation_rate, results, index):
    population_ranked = rank_routes(current_generation)
    matingpool = selection(population_ranked, elite_count)
    children = breed_population(matingpool, elite_count)
    next_generation = mutate_population(children, mutation_rate)
    results[index] = next_generation
```

## Conventional Genetic Algorithm

```python
def conventional_genetic_alrorithm(population, pop_size, elite_count, mutation_rate, generations):
    pop = initialize_population(pop_size, population)
    #print("Initial distance: " + str( rank_routes(pop)[0][1] ) )
    
    for i in range(0, generations):
        pop = next_generation(pop, elite_count, mutation_rate)
        #print("Generation distance" + str(i) + ": " + str( rank_routes(pop)[0][1] ) )
    
    #print("Final distance: " + str( rank_routes(pop)[0][1] ) )
    
    bestRoute = rank_routes(pop)[0]
    
    return bestRoute
```

```python
def conventional_genetic_alrorithm_plot(population, pop_size, elite_count, mutation_rate, generations):
    pop = initialize_population(pop_size, population)
    progress = []
    progress.append(rank_routes(pop)[0][1])
    
    for i in range(0, generations):
        pop = next_generation(pop, elite_count, mutation_rate)
        progress.append(rank_routes(pop)[0][1])
    
    #print("Final distance: " + str( rank_routes(pop)[0][1] ) )
    
    
    plt.plot(progress)
    plt.ylabel('Distance')
    plt.xlabel('Generation')
    plt.title('Conventional GA')
    plt.show()
    
    
    bestRoute = rank_routes(pop)[0]
    
    return bestRoute
```

### Tread Class for return of results

```python
# https://stackoverflow.com/questions/6893968/how-to-get-the-return-value-from-a-thread-in-python?answertab=votes#tab-top

class ThreadWithReturnValue(Thread):
    def __init__(self, group=None, target=None, name=None,
                 args=(), kwargs={}, Verbose=None):
        Thread.__init__(self, group, target, name, args, kwargs)
        self._return = None
    def run(self):
        if self._target is not None:
            self._return = self._target(*self._args,
                                                **self._kwargs)
    def join(self, *args):
        Thread.join(self, *args)
        return self._return
```

### Migration function for moving individuals bewtween sub populations

```python
def migrate(populations, migration_rate, migration_count):
    sub_population_count = len(populations)
            
    if sub_population_count == 1: # cannot migrate a single population
        return
    
    if random.random() < migration_rate:

        if sub_population_count == 2: # only do a single swap when there are only two in the population
            
            for i in range(migration_count):
                temp = populations[1][i]
                populations[1][i] = populations[0][i]
                populations[0][i] = temp
                
            return
        
        else:
            unassigned = []

            for i in range(sub_population_count):

                fro = i
                to = i + 1

                if to == sub_population_count:
                    to = 0
                
                if fro == 0:
                    for i in range(migration_count):
                        unassigned.append(populations[to][i])
                        populations[to][i] = populations[fro][i]
                else:
                    temp = []
                    for i in range(migration_count):
                        temp.append(populations[to][i])
                        populations[to][i] = unassigned[i]
                    unassigned = temp
```

## Parallel Genetic Algorithm

```python
def parrallel_genetic_alrorithm(population, pop_size, elite_count, mutation_rate, generations, sub_population_count, migration_rate, migration_count):
    
    thread_count = sub_population_count
    threads = [None] * thread_count
    results = [None] * thread_count
    
    elite_counts = [elite_count for _ in range(thread_count)]
    mutation_rates = [mutation_rate for _ in range(thread_count)]
    
    sub_population_size = round(pop_size/thread_count)
    
    populations = [initialize_population(sub_population_size, population) for _ in range(thread_count)]
    
    for g in range(generations):
        
        for i in range(thread_count):
            threads[i] = Thread(target=next_generation_for_treads, args=(populations[i], elite_counts[i], mutation_rates[i], results, i))
            threads[i].start()

        for i in range(thread_count):
            threads[i].join()
        
        migrate(results, migration_rate, migration_count)
        
        populations = results
    
    bestroute = []
    bestfitness = float('-inf')
    for r in results:
        if r[0][2] > bestfitness: 
            bestfitness = r[0][2]
            bestroute = r[0]
    
    return bestroute
                                
```
